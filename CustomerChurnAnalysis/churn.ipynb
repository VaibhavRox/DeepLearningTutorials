## Customer Churn Prediction using Artifical Neural Network

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
#Load the data
df=pd.read_csv("customer_churn.csv")
df.sample(5)
```

```python
#Drop CustomerID column
df.drop("customerID",axis=1,inplace=True)
```

```python
df.dtypes
```

```python
#We see total charges is an object type, we need to convert it to float
df.TotalCharges.values
```

```python
#Its string type, convert it to numbers
pd.to_numeric(df.TotalCharges)
```

```python
#There are some blank values in TotalCharges, we need to remove them
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
#Remove the rows with blank values
df1=df[df.TotalCharges!=' ']
df1.shape
```

```python
df1.dtypes
```

```python
df1.TotalCharges=pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges.values
```

```python
df1[df1.Churn=="No"]
```

```python
#Data Visualization
tenure_churn_no=df1[df1.Churn=="No"].tenure
tenure_churn_yes=df1[df1.Churn=="Yes"].tenure

plt.xlabel("Tenure (in months)")
plt.ylabel("Number of Customers")
plt.title("Customer Churn Prediction")

plt.hist([tenure_churn_yes, tenure_churn_no],rwidth=0.95,color=['green','red'],label=['Churn=Yes','Churn=No'])
plt.legend()
```

```python
#Do same plot for MonthlyCharges
mc_churn_no=df1[df1.Churn=="No"].MonthlyCharges
mc_churn_yes=df1[df1.Churn=="Yes"].MonthlyCharges

plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.title("Customer Churn Prediction")
plt.hist([mc_churn_yes, mc_churn_no],rwidth=0.95,color=['green','red'],label=['Churn=Yes','Churn=No'])
plt.legend()
```

Many of the columns are yes,no etc. Print out the unique values in object columns to see data values

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtype=='object':
            print(f"{column}:{df[column].unique()}")
```

```python
print_unique_col_values(df1)
```

```python
#Some columns have no phone service, some have no internet service. We need to convert these to no
df1.replace('No internet service', 'No', inplace=True)
df1.replace('No phone service', 'No', inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
#Convert yes/no columns to 1/0
yes_no_cols=['Partner', 'Dependents', 'PhoneService', 'MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
             'OnlineBackup','TechSupport','DeviceProtection','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for column in yes_no_cols:
    df1[column].replace({'Yes':1,'No':0},inplace=True)
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.gender.unique()
```

One hot Encoding for categorical columns

```python
df2=pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'],dtype=int)
df2.columns
```

```python
df2.sample(5)
```

```python
df2.dtypes
```

```python
#Scale the data
cols_to_scale=['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}:{df2[col].unique()}')
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X=df2.drop('Churn',axis='columns')
Y=df2['Churn']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

# Build a model (ANN) in tensorflow/keras

```python
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Dense(26,input_shape=(26,),activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=100)
```

```python
model.evaluate(X_test, Y_test)
```

```python
yp=model.predict(X_test)
yp[:5]
```

```python
y_pred=[]
for element in yp:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
Y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(Y_test, y_pred))
```

```python
import seaborn as sns
cm=tf.math.confusion_matrix(labels=Y_test,predictions=y_pred)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
